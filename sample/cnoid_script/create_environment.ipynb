# 環境ファイルの書き方

https://github.com/IRSL-tut/irsl_sim_environments/blob/main/cnoid/world/depot_world.yaml
```yaml
world:
  World:
    name: DepotWorld
    draw_grid: False
  Camera:
    lookEye: [-5.5, -0.0, 2.0]
    lookUp: [0.0, 0.0, 1.0]
    lookAtCenter: [-2.5, -0.0, 0.7]
    fov: 0.75

objects:
  -
    model: 'package://irsl_sim_environments/cnoid/depot/depot.body'
    name: 'Depot'
    fix: True
  -
    model: 'package://irsl_sim_environments/cnoid/dining_table/dining_table.body'
    name: 'Table'
    initial_coords: { pos: [-2.5, 0.0, 0.0] }
    fix: True
```

## world （トップアイテム）
以下のキーワードの設定ができる
- World
- Camera

### Worldの設定
- name: WorldItemの名前
- draw_grid: gridを描画するかどうか
- background: 背景色(未実装）

### Cameraの設定
- lookEye: カメラの原点
- lookUp: 画像の上方向
- lookAtCener: カメラの光軸が通る点

# objects （トップアイテム）
objectアイテムのリストを持つ

## objectアイテムの設定
- model: bodyファイル
- name: objectの名前
- initial_coords: 初期位置 (irsl_choreonoid.robot_util.make_coordinatesで読める形式)
- fix: Trueを設定すると固定環境になる
